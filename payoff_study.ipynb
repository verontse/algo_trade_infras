```python
import numpy as np
import seaborn as sn
```

```python
class Portfolio():
    def __init__(self, name):
        self.name = name
        self.stock_lst = []
        self.option_lst = []
    
    def add_options(self, option):
        self.option_lst.append(option)

    def add_stocks(self, stock):
        self.stock_lst.append(stock)
        
    def plot_payoffs(self):
        pass
```

```python
class Option():
    def __init__(self, strike, premium, expiry=None):
        self.strike = strike
        self.premium = premium
        self.expiry = expiry
```

```python
a = Option(strike=100, premium=1)
b = Option(strike=80, premium=1)
options_lst = [a, b]
interval = 5
plot_max = round(max(a.strike, b.strike) * 1.5)
x = np.linspace(0, plot_max, plot_max//interval+1)

temp_payoff = np.zeros(len(x))

for option in options_lst:
    payoff = [i - option.strike - option.premium if i > option.strike else - option.premium for i in x]
    temp_payoff += payoff
sn.lineplot(data=temp_payoff, x=x)
```
